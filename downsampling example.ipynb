```python
import numpy as np 
import chainer.functions as F 
import chainer.links as L
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# ground truth class labeled image  
im_shape = 240
channels= 1
gt_class = np.zeros((1,channels,im_shape,im_shape))
```

```python

gt_display = gt_class.reshape((im_shape,im_shape, channels))
gt_display[100:200, 100:200, :] =1 
gt_class = gt_display.reshape((1,channels,im_shape, im_shape))

plt.figure()
plt.imshow(gt_display[:,:,0])
plt.title('Full image')
plt.show()
ones_list = np.where(gt_display==1)
```

```python
down_ones_list = np.divide(ones_list,4) 
down_ones_list = down_ones_list.astype(int)
```

```python

h = F.max_pooling_2d(gt_class, 2, stride=2 )
h2 = F.max_pooling_2d(h, 2, stride=2 )
```

```python
gt_display = h2.reshape((h2.shape[2],h2.shape[2]))

plt.figure()
plt.imshow(gt_display.data)
plt.title('downsampled image')
plt.show()
```

```python
down_sampled_list  = np.where(gt_display.data==1)
```
